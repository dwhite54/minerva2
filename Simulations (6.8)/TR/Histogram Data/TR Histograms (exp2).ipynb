```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1% and 1:1

```python
fig, axs = plt.subplots(3, 3, sharex=True, sharey=True)
fig.set_figheight(14)
fig.set_figwidth(12)
#fig.suptitle("Echo intensity as a function of repetition")
conds=['Rhythm','Pitch','Rhythm+Pitch']

for i, TONE_RHYTHM_RATIO in enumerate(['11','23','37']):
    for j, NOISE_RATIO in enumerate(['.01', '.1','.2']):
        data = pd.read_csv('N{}_{}.csv'.format(NOISE_RATIO, TONE_RHYTHM_RATIO))     
        for cond in conds:
            subset = data[data['condition']==cond]  
            sns.distplot(subset['echo'], hist=False,kde=True,
                        kde_kws={'shade':True,'linewidth':3},
                        label=cond, ax=axs[i,j])
            
        axs[i, j].set_title("{:.0f}% Noise, {}:{} Ratio".format(
            float(NOISE_RATIO)*100, TONE_RHYTHM_RATIO[0], TONE_RHYTHM_RATIO[1]))
        axs[i, j].set_xlabel("Echo Intensity")
        axs[i, j].set_ylabel("Density")
        axs[i, j].set_ylim(0, 60000)
        axs[i, j].set_xlim(-1e-4, 0.001)
        axs[i, j].grid(False)
        axs[i, j].legend(loc=1, prop={'size':13.5})
        axs[i, j].label_outer()
            

plt.tight_layout()
plt.savefig("TR_Hist.png")
plt.show()
```
