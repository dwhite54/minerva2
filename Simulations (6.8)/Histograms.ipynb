```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_color_codes('bright')

data_N01_11 = pd.read_csv('Hist_N.01_1.1.csv')
```

```python
sns.distplot(data_N01_11['echo'], hist=True, kde=False)
```

```python
sns.set_color_codes('bright')
cond=['Rhythm','Pitch','Rhythm+Pitch']

for conds in cond:
    subset = data_N01_11[data_N01_11['condition']==conds]
    
    sns.distplot(subset['echo'], hist=False,kde=True,
                kde_kws={'shade':True,'linewidth':3},
                label=conds)
    plt.title('1% Noise; 1:1 Ratio')
    plt.xlabel('Echo Intensity')
    plt.ylabel('Density')
```
